```python
#Taking P,T and R from user and calculate SI using function
def calculate_si(P, T, R):
    SI = (P * T * R) / 100
    return SI

P = float(input("Enter Principal Amount (P): "))
T = float(input("Enter Time (T in years): "))
R = float(input("Enter Rate of Interest (R): "))

result = calculate_si(P, T, R)


print(f"\nSimple Interest (SI) = {result}")

```
